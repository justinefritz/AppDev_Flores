**1. Class vs. Object vs. Instance:**

**-Class:** A blueprint or template for creating objects.

**-Object (or Instance):** Created from a class and represents a specific entity with its own unique characteristics.

**-Instance:** Just another name for an object.

**2. Method vs. Function:**

**-Method:** A function defined inside a class, specifically designed to work with the object's data.

**-Function:** A block of code that performs a specific task, often standalone.

**3. Attribute vs. Variable:**

**-Attribute:** Data attached to an object, representing its state.

**-Variable:** A named container for storing data, often used within functions or methods.

**4. Procedural vs. Object-Oriented Programming:**

**-Procedural Programming:** Focuses on writing procedures or routines to perform tasks. Data and functions are separate.

**-Object-Oriented Programming (OOP):** Focuses on creating objects that contain both data and functions (methods). Data and functions are bundled together into objects.

**5. Advantage of Using a Class Over Modules:**

**-Encapsulation:** Classes bundle data and functions together, promoting data privacy and integrity.

**-Inheritance:** Classes can inherit features from other classes, promoting code reuse.

**-Polymorphism:** Allows objects of different classes to be treated uniformly, promoting flexibility.

**-Instance Specificity:** Classes allow creating multiple unique instances, each with its own characteristics.
